```python
df.dropna(subset=["TweetText"],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df['TweetText'] = df['TweetText'].str.replace('im', 'i am')
df['TweetText'] = df['TweetText'].str.replace('dont', 'do not')
```

```python
import string
string.punctuation
```

```python
punctuations = string.punctuation

df['new'] = df['TweetText'].apply(lambda x: ''.join(char for char in x if char not in punctuations))
pd.set_option('display.max_colwidth',300)
df.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
df['nostopwords'] = df['new'].apply(lambda x: ' '.join(word for word in x.split() if word.lower() not in stop_words))
df.head()
```

```python
df['lowertext'] = df['new'].apply(lambda x: x.lower())
df.head(10)
```

```python
import re 
```

```python
def tokenize(text):
    return re.split('\W+', text)

df['tokenization'] = df['lowertext'].apply(tokenize)

df.head()
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(words):
    return [ps.stem(word) for word in words]

df['stemming'] = df['tokenization'].apply(lambda x: stem_words(x))
df.head()
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

# Function for lemmatizing a list of words
def lemmatize_words(words):
    return [wnl.lemmatize(word) for word in words]


df['lemmatization'] = df['stemming'].apply(lambda x: lemmatize_words(x))
df.head(10)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
from textblob import TextBlob

def pos_tagging(text):
    blob = TextBlob(text)
    return blob.tags
df_subset = df.head(100)
df_subset['pos-tagging'] = df_subset['lemmatization'].apply(lambda words: pos_tagging(" ".join(words)))
df_subset['pos-tagging']
```

```python
df_subset.head(10)
```

```python
N = len(df_subset)
def doc_freq(word):
    df = sum(1 for text in df_subset['lemmatization'] if word in text)
    return df

def calc_tf_idf(doc, token):
    # calculate the tf-idf score for the given token in the given document
    words_count = len(df_subset['lemmatization'][doc])
    counter = dict(nltk.FreqDist(df_subset['lemmatization'][doc]))
    tf = counter[token]/words_count
    df = doc_freq(token)
    idf = np.log(N/(df+1))
    tf_idf = tf*idf
    return tf_idf

# create a list to store the tf-idf scores for each row
tf_idf_list = []

for doc in range(N):
    # create a dictionary to store the tf-idf scores for the tokens in the current row
    tf_idf_dict = {}
    # iterate over each token in the current row
    for token in df_subset['lemmatization'][doc]:
        # calculate the tf-idf score for the current token in the current row
        tf_idf = calc_tf_idf(doc=doc, token=token)
        # add the tf-idf score to the dictionary
        tf_idf_dict[token] = tf_idf
    # add the dictionary of tf-idf scores for the current row to the list
    tf_idf_list.append(tf_idf_dict)

df_subset['tf_idf'] = tf_idf_list
```

```python
from nltk.stem import PorterStemmer
from spellchecker import SpellChecker

ps = PorterStemmer()
spell = SpellChecker()

def stem_words_and_correct_spelling(words):
    stemmed_words = [ps.stem(word) for word in words]
    corrected_words = [spell.correction(word) for word in stemmed_words]
    return corrected_words

df_subset['Corrected_AfterStemming'] = df_subset['tokenization'].apply(stem_words_and_correct_spelling)
```

```python
df_subset
```

```python
N = len(df_subset)
def doc_freq(word):
    df = sum(1 for text in df_subset['Corrected_AfterStemming'] if word in text)
    return df

def calc_tf_idf(doc, token):
    # calculate the tf-idf score for the given token in the given document
    words_count = len(df_subset['Corrected_AfterStemming'][doc])
    counter = dict(nltk.FreqDist(df_subset['Corrected_AfterStemming'][doc]))
    tf = counter[token]/words_count
    df = doc_freq(token)
    idf = np.log(N/(df+1))
    tf_idf = tf*idf
    return tf_idf

# create a list to store the tf-idf scores for each row
tf_idf_list = []

for doc in range(N):
    # create a dictionary to store the tf-idf scores for the tokens in the current row
    tf_idf_dict = {}
    # iterate over each token in the current row
    for token in df_subset['Corrected_AfterStemming'][doc]:
        # calculate the tf-idf score for the current token in the current row
        tf_idf = calc_tf_idf(doc=doc, token=token)
        # add the tf-idf score to the dictionary
        tf_idf_dict[token] = tf_idf
    # add the dictionary of tf-idf scores for the current row to the list
    tf_idf_list.append(tf_idf_dict)

df_subset['tfidf_AfterCorrection'] = tf_idf_list
```
